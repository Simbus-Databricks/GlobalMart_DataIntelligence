```python
spark.sql("USE CATALOG globalmart_dev")
print("Active catalog: globalmart_dev")
```

```python
%sql
SELECT 'customers'    AS entity, COUNT(*) AS rows FROM globalmart_dev.bronze.customers
UNION ALL
SELECT 'orders',       COUNT(*) FROM globalmart_dev.bronze.orders
UNION ALL
SELECT 'transactions', COUNT(*) FROM globalmart_dev.bronze.transactions
UNION ALL
SELECT 'returns',      COUNT(*) FROM globalmart_dev.bronze.returns
UNION ALL
SELECT 'products',     COUNT(*) FROM globalmart_dev.bronze.products
UNION ALL
SELECT 'vendors',      COUNT(*) FROM globalmart_dev.bronze.vendors
```

```python
%sql
SELECT
    _source_file_name,
    source_region,
    COUNT(*) AS rows
FROM globalmart_dev.bronze.customers
GROUP BY _source_file_name, source_region
ORDER BY source_region
```

```python
%sql
SELECT
    source_region,
    COUNT(customer_id)          AS has_customer_id,
    COUNT(CustomerID)           AS has_CustomerID,
    COUNT(cust_id)              AS has_cust_id,
    COUNT(customer_identifier)  AS has_customer_identifier
FROM globalmart_dev.bronze.customers
GROUP BY source_region
ORDER BY source_region
 
```

```python
%sql
SELECT
    source_region,
    COUNT(customer_email)  AS has_customer_email,
    COUNT(email_address)   AS has_email_address
FROM globalmart_dev.bronze.customers
GROUP BY source_region
ORDER BY source_region
```

```python
%sql
SELECT
    _source_file_name,
    source_region,
    _rescued_data
FROM globalmart_dev.bronze.customers
WHERE _rescued_data IS NOT NULL
LIMIT 20
```

```python
%sql
SELECT
    _source_file_name,
    source_region,
    _rescued_data
FROM globalmart_dev.bronze.transactions
WHERE _rescued_data IS NOT NULL
LIMIT 20
```

```python
%sql
SELECT
    _source_file_name,
    source_region,
    COUNT(*) AS rows
FROM globalmart_dev.bronze.transactions
GROUP BY _source_file_name, source_region
ORDER BY source_region
```

```python
%sql
SELECT
    _source_file_name,
    source_region,
    COUNT(order_id)       AS has_order_id,
    COUNT(OrderId)        AS has_OrderId,
    COUNT(return_reason)  AS has_return_reason,
    COUNT(reason)         AS has_reason,
    COUNT(refund_amount)  AS has_refund_amount,
    COUNT(amount)         AS has_amount,
    COUNT(return_status)  AS has_return_status,
    COUNT(status)         AS has_status
FROM globalmart_dev.bronze.returns
GROUP BY _source_file_name, source_region
```

```python
%sql
SELECT 'customers'    AS entity, source_region, COUNT(*) AS rows FROM globalmart_dev.bronze.customers    GROUP BY source_region
UNION ALL
SELECT 'orders',       source_region, COUNT(*) FROM globalmart_dev.bronze.orders       GROUP BY source_region
UNION ALL
SELECT 'transactions', source_region, COUNT(*) FROM globalmart_dev.bronze.transactions GROUP BY source_region
UNION ALL
SELECT 'returns',      source_region, COUNT(*) FROM globalmart_dev.bronze.returns      GROUP BY source_region
UNION ALL
SELECT 'products',     source_region, COUNT(*) FROM globalmart_dev.bronze.products     GROUP BY source_region
UNION ALL
SELECT 'vendors',      source_region, COUNT(*) FROM globalmart_dev.bronze.vendors      GROUP BY source_region
ORDER BY entity, source_region
```

```python
%sql
SELECT
    'customers' AS entity,
    COUNT(*) AS total_rows,
    COUNT(_source_file_path)     AS has_file_path,
    COUNT(_source_file_name)     AS has_file_name,
    COUNT(_source_modified_time) AS has_modified_time,
    COUNT(_ingested_at)          AS has_ingested_at
FROM globalmart_dev.bronze.customers
UNION ALL
SELECT 'orders', COUNT(*), COUNT(_source_file_path), COUNT(_source_file_name), COUNT(_source_modified_time), COUNT(_ingested_at) FROM globalmart_dev.bronze.orders
UNION ALL
SELECT 'transactions', COUNT(*), COUNT(_source_file_path), COUNT(_source_file_name), COUNT(_source_modified_time), COUNT(_ingested_at) FROM globalmart_dev.bronze.transactions
UNION ALL
SELECT 'returns', COUNT(*), COUNT(_source_file_path), COUNT(_source_file_name), COUNT(_source_modified_time), COUNT(_ingested_at) FROM globalmart_dev.bronze.returns
UNION ALL
SELECT 'products', COUNT(*), COUNT(_source_file_path), COUNT(_source_file_name), COUNT(_source_modified_time), COUNT(_ingested_at) FROM globalmart_dev.bronze.products
UNION ALL
SELECT 'vendors', COUNT(*), COUNT(_source_file_path), COUNT(_source_file_name), COUNT(_source_modified_time), COUNT(_ingested_at) FROM globalmart_dev.bronze.vendors
```
